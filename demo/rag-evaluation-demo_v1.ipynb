```python
def append_to_csv(step_name, scores, filename='rag_evaluation_results.csv'):
    file_exists = os.path.isfile(filename)
    with open(filename, 'a', newline='') as f:
        writer = csv.writer(f)
        if not file_exists:
            writer.writerow(['step', 'precision', 'recall', 'f1'])
        writer.writerow([step_name, scores['precision'], scores['recall'], scores['f1']])
```
